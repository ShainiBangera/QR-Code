```python
import qrcode
from IPython.display import Image, display
from google.colab import files

#taking input from user
txt = input("Enter text or URL for QR Code:")

#Creating QR code
qr = qrcode.QRCode(
    version = 1, #size of qr code
    box_size = 10,
    border = 5
)

qr.add_data(txt)
qr.make(fit = True)

#Generate image
img = qr.make_image(fill = "black", back_color = "white")
img.save("qrcode.png")

print("QR Code generated successfully!")

# Display image
display(Image(filename="qrcode.png"))
```
